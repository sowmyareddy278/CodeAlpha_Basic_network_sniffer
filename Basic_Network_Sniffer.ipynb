```python
from scapy.all import *
```

```python
def packet_analyzer(packet):
  if packet.haslayer(IP):
    sour_ip = packet[IP].src
    dest_ip = packet[IP].dst
    protocol = packet[IP].proto
    print(f"IP Packet: {sour_ip} --> {dest_ip} Protocol: {protocol}")

  if packet.haslayer(TCP):
    sour_port = packet[TCP].sport
    dest_port = packet[TCP].dport
    print(f"TCP Packet: {sour_port} --> {dest_port} Payload: {str(packet[TCP].payload)}")

  if packet.haslayer(UDP):
    sour_port = packet[UDP].sourceport
    dest_port = packet[UDP].destport
    print(f"UDP Packet: {sour_port} --> {dst_port} Payload: {str(packet[UDP].payload)}")

sniff(prn=packet_analyzer, store=0)
```
